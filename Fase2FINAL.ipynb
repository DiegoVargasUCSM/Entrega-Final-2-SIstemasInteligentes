```python

# 1. IMPORTAR LIBRERÍAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

from google.colab import files

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 2. CARGAR ARCHIVO CSV
print("Sube tu archivo ventas_productos.csv")
uploaded = files.upload()

nombre_archivo = list(uploaded.keys())[0]

df = pd.read_csv(nombre_archivo)

print("\nArchivo cargado correctamente:")
print(nombre_archivo)

print("\nPrimeras filas del dataset:")
display(df.head())

# 3. INFORMACIÓN GENERAL
print("\nCantidad de filas y columnas:")
print(df.shape)

print("\nInformación del dataset:")
print(df.info())

print("\nValores nulos por columna:")
print(df.isnull().sum())

print("\nEstadísticas generales:")
display(df.describe())

# 4. VALIDAR MÍNIMO 1000 REGISTROS
cantidad_registros = len(df)

print("\nValidación de cantidad de registros:")

if cantidad_registros >= 1000:
    print("El dataset cumple con el requisito mínimo.")
    print("Cantidad de registros:", cantidad_registros)
else:
    print("El dataset NO cumple con el requisito mínimo.")
    print("Cantidad de registros:", cantidad_registros)

# 5. LIMPIEZA DE DATOS
print("\nIniciando limpieza de datos...")

# Eliminar duplicados
df = df.drop_duplicates()

# Eliminar registros con valores nulos
df = df.dropna()

print("Dataset después de limpieza:")
print(df.shape)

print("\nValores nulos después de limpieza:")
print(df.isnull().sum())

# 6. DEFINIR VARIABLE OBJETIVO
objetivo = "Ventas"

if objetivo not in df.columns:
    raise ValueError("No existe la columna 'Ventas' en el CSV.")

# Eliminar ID si existe, porque no sirve para entrenar
columnas_eliminar = [objetivo]

if "ID" in df.columns:
    columnas_eliminar.append("ID")

X = df.drop(columns=columnas_eliminar)
y = df[objetivo]

print("\nVariables de entrada X:")
print(X.columns.tolist())

print("\nVariable objetivo y:")
print(objetivo)

# 7. IDENTIFICAR COLUMNAS
columnas_numericas = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
columnas_categoricas = X.select_dtypes(include=["object"]).columns.tolist()

print("\nColumnas numéricas:")
print(columnas_numericas)

print("\nColumnas categóricas:")
print(columnas_categoricas)

# 8. DIVIDIR DATOS
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\nDatos de entrenamiento:")
print(X_train.shape)

print("\nDatos de prueba:")
print(X_test.shape)

# 9. PREPROCESAMIENTO
preprocesador = ColumnTransformer(
    transformers=[
        ("numericas", StandardScaler(), columnas_numericas),
        ("categoricas", OneHotEncoder(handle_unknown="ignore"), columnas_categoricas)
    ]
)

# 10. CREAR MODELOS
modelos = {
    "KNN Regressor": KNeighborsRegressor(n_neighbors=5),
    "SVM Regressor": SVR(kernel="rbf"),
    "Árbol de Decisión": DecisionTreeRegressor(random_state=42)
}

# 11. ENTRENAR Y EVALUAR MODELOS
resultados = []
modelos_entrenados = {}

print("\nEntrenando modelos...\n")

for nombre, modelo in modelos.items():

    pipeline = Pipeline(steps=[
        ("preprocesador", preprocesador),
        ("modelo", modelo)
    ])

    # Entrenar modelo
    pipeline.fit(X_train, y_train)

    # Realizar predicciones
    y_pred = pipeline.predict(X_test)

    # Calcular métricas
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    resultados.append({
        "Modelo": nombre,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    })

    modelos_entrenados[nombre] = pipeline

    print("Modelo:", nombre)
    print("MAE:", round(mae, 2))
    print("RMSE:", round(rmse, 2))
    print("R2:", round(r2, 4))
    print("-----------------------------------")

# 12. TABLA COMPARATIVA
df_resultados = pd.DataFrame(resultados)
df_resultados = df_resultados.sort_values(by="R2", ascending=False)

print("\nTabla comparativa de modelos:")
display(df_resultados)

# 13. SELECCIONAR MEJOR MODELO
mejor_modelo_nombre = df_resultados.iloc[0]["Modelo"]
mejor_modelo = modelos_entrenados[mejor_modelo_nombre]

print("\nMejor modelo seleccionado:")
print(mejor_modelo_nombre)

print("\nMétricas del mejor modelo:")
display(df_resultados.iloc[0])

# 14. GRÁFICA: DISTRIBUCIÓN DE VENTAS
plt.figure(figsize=(8, 5))
plt.hist(df["Ventas"], bins=30)
plt.title("Distribución de Ventas")
plt.xlabel("Cantidad de ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# 15. GRÁFICA: PRECIO VS VENTAS
if "Precio" in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df["Precio"], df["Ventas"])
    plt.title("Relación entre Precio y Ventas")
    plt.xlabel("Precio del producto")
    plt.ylabel("Ventas")
    plt.grid(True)
    plt.show()

# 16. GRÁFICA: DESCUENTO VS VENTAS
if "Descuento" in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df["Descuento"], df["Ventas"])
    plt.title("Relación entre Descuento y Ventas")
    plt.xlabel("Descuento (%)")
    plt.ylabel("Ventas")
    plt.grid(True)
    plt.show()

# 17. GRÁFICA: STOCK VS VENTAS
if "Stock" in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df["Stock"], df["Ventas"])
    plt.title("Relación entre Stock y Ventas")
    plt.xlabel("Stock disponible")
    plt.ylabel("Ventas")
    plt.grid(True)
    plt.show()

# 18. GRÁFICA: COMPARACIÓN POR R2
plt.figure(figsize=(8, 5))
plt.bar(df_resultados["Modelo"], df_resultados["R2"])
plt.title("Comparación de Modelos según R2")
plt.xlabel("Modelo")
plt.ylabel("R2")
plt.xticks(rotation=20)
plt.grid(True)
plt.show()

# 19. GRÁFICA: COMPARACIÓN POR RMSE
plt.figure(figsize=(8, 5))
plt.bar(df_resultados["Modelo"], df_resultados["RMSE"])
plt.title("Comparación de Modelos según RMSE")
plt.xlabel("Modelo")
plt.ylabel("RMSE")
plt.xticks(rotation=20)
plt.grid(True)
plt.show()

# 20. PREDICCIÓN DE PRUEBA
nuevo_producto = pd.DataFrame([{
    "Producto": "Laptop",
    "Categoria": "Tecnologia",
    "Precio": 2500.00,
    "Descuento": 15,
    "Stock": 120,
    "Publicidad": "Si",
    "Calificacion": 4.5,
    "Visitas_Web": 5000,
    "Mes": "Mayo",
    "Dia_Semana": "Lunes",
    "Temporada": "Normal"
}])

# Ordenar columnas igual que en entrenamiento
nuevo_producto = nuevo_producto[X.columns]

prediccion = mejor_modelo.predict(nuevo_producto)

print("\nPredicción de ventas para nuevo producto:")
print(round(prediccion[0]), "unidades")

# 21. GUARDAR MEJOR MODELO
paquete_modelo = {
    "modelo": mejor_modelo,
    "nombre_modelo": mejor_modelo_nombre,
    "columnas_entrada": X.columns.tolist(),
    "metricas": df_resultados
}

joblib.dump(paquete_modelo, "modelo_prediccion_ventas.pkl")

print("\nModelo guardado correctamente:")
print("modelo_prediccion_ventas.pkl")

# 22. GUARDAR MÉTRICAS
df_resultados.to_csv("metricas_modelos.csv", index=False)

print("\nMétricas guardadas correctamente:")
print("metricas_modelos.csv")

# 23. DESCARGAR ARCHIVOS
files.download("modelo_prediccion_ventas.pkl")
files.download("metricas_modelos.csv")
```
